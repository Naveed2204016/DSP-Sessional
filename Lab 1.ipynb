```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
t=np.arange(0,1,0.001)
amp=1*np.sin(2*np.pi*1*t)
plt.plot(t,amp)
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("sinusoidal wave")
plt.show()
```

```python
t=np.arange(0,1,0.001)
amp=1*np.cos(2*np.pi*1*t)
plt.plot(t,amp)
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("cosine wave")
plt.show()
```

```python
t=np.arange(0,1,0.01)
amp=1*np.sin(2*np.pi*1*t)
plt.plot(t,amp)
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("sinusoidal wave")
plt.show()
```

```python
t=np.arange(0,1,0.1)
amp=1*np.sin(2*np.pi*1*t)
plt.stem(t,amp)
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("sinusoidal wave")
plt.show()
```
