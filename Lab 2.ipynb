```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
f1=10
f2=50
duration=1
Fs=40
```

```python
time=np.arange(0,duration,(1/Fs))
amp1=np.sin(2*np.pi*f1*time)
amp2=np.sin(2*np.pi*f2*time)
plt.plot(time,amp1)
plt.show()
plt.plot(time,amp2)
plt.show()
```
